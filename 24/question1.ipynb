```python
from threading import Thread
from time import time
from moviepy import editor
```

```python



videopaths=["input/6minute1.mp4","input/6minute2.mp4","input/6minute3.mp4","input/6minute4.mp4","input/6minute5.mp4"]
audiopaths=["output/6minute1.mp3","output/6minute2.mp3","output/6minute3.mp3","output/6minute4.mp3","output/6minute5.mp3"]



def convert(path_video,path_audio):
    video = editor.VideoFileClip(path_video)
    video.audio.write_audiofile(path_audio)

start_time=time()

for i in range(5):
    convert(videopaths[i],audiopaths[i])

end_time=time()

process_time=end_time-start_time

print(process_time)

# process time   =====>   30.3709774017334
```

```python
videopaths=["input/6minute1.mp4","input/6minute2.mp4","input/6minute3.mp4","input/6minute4.mp4","input/6minute5.mp4"]
audiopaths=["output/6minute1.mp3","output/6minute2.mp3","output/6minute3.mp3","output/6minute4.mp3","output/6minute5.mp3"]



def convert(path_video,path_audio):
    video = editor.VideoFileClip(path_video)
    video.audio.write_audiofile(path_audio)

start_time=time()

threads=[]
for i in range(5):
    threads.append(Thread(target=convert,args=[videopaths[i],audiopaths[i]]))      

for t in threads:
    t.start()

for t in threads:
    t.join()

end_time=time()

process_time=end_time-start_time

print(process_time)


# process time   =====>   13.974942684173584
```
