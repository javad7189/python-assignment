```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(121) 

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

balloons = pd.DataFrame({
'width': np.random.normal(40,10,400),
'lenght':np.random.normal(50,35,400),
'weight':np.random.normal(10,3,400)}
)

melon = pd.DataFrame({
'width':np.random.normal(40,10,400),
'lenght':np.random.normal(50,25,400),
'weight':np.random.normal(4000,1000,400)})


balloons.to_csv("output/balloons.csv")
melon.to_csv("output/melon.csv")

ax.scatter(balloons["weight"],balloons["lenght"],balloons["width"], marker="o")
ax.scatter(melon["weight"],melon["lenght"],melon["width"] ,marker="^")


ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
